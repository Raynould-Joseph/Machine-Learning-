#ii)Convert the given images in to different formats

# to png format

```python
import cv2
img = cv2.imread("/content/sample_data/img1.jpg")
cv2.imwrite('/content/sample_data/output.png', img, [cv2.IMWRITE_PNG_COMPRESSION])
```

# to pam format

```python
import cv2
img = cv2.imread("/content/sample_data/img1.jpg")
cv2.imwrite('/content/sample_data/output.pam', img, [cv2.IMWRITE_PAM_FORMAT_GRAYSCALE 
])
```
